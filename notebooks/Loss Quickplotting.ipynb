# Imports

```python
import pickle
import torch
import matplotlib.pyplot as plt
```

# Predictor

```python
!ls ../models/checkpoints/LSTM_E4_H64
```

```python
checkpoint_num = 100000


with open('../models/checkpoints/LogP Predictor/losses-{}.pkl'.format(checkpoint_num), 'rb') as f:
    losses = pickle.load(f)
```

```python
plt.plot(losses)
plt.xlabel("Iteration # (in thousands)")
plt.ylabel("Cross Entropy Loss")
plt.title("Predictor Loss")
plt.savefig("predictor_loss.png")
```
